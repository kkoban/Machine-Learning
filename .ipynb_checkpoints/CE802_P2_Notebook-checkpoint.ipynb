### Part A

```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('CE802_P2_Data.csv')
train.head(20)
```

```python
train.info()
```

```python
sns.countplot(train['Class'])
```

```python
train.drop('F21', axis = 1, inplace = True)
train.info()
```

```python
train['F8'].plot.box(grid='True')
```

```python
# Position of the Outlier
x = np.where(train['F8']<-60000)
train.drop(x[0], axis = 0, inplace = True)
train.info()
```

```python
train['F8'].plot.box(grid='True')
```

```python
train.plot.box(grid='True')
```

```python
from sklearn.model_selection import train_test_split
X = train.drop('Class',axis=1)
y = train[['Class']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state=42)
```

```python
X_test.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier(criterion="gini",max_depth=5)   
#The higher value of maximum depth causes overfitting, and a lower value causes underfitting (Source).
dtc_model.fit(X_train,y_train)
```

```python
sklearn.tree.plot_tree(dtc_model)
```

```python
y_predict = dtc_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,y_predict)
```
